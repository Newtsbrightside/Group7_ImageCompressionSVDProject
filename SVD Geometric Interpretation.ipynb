```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2,1],
              [1,2]])
U, S, Vt = np.linalg.svd(A)

# unit circle
theta = np.linspace(0, 2*np.pi, 200)
circle = np.array([np.cos(theta), np.sin(theta)])

V_circle = Vt @ circle            # rotate
Sigma = np.zeros((2, 2))
Sigma[0, 0] = S[0]
Sigma[1, 1] = S[1]
Sigma_circle = Sigma @ V_circle   # stretch
U_circle = U @ Sigma_circle       # rotate


plt.figure(figsize=(6, 6))
plt.plot(circle[0, :], circle[1, :], 'k--', label='Unit circle')
plt.plot(V_circle[0, :], V_circle[1, :], 'g', label='Vᵀ·x')
plt.plot(Sigma_circle[0, :], Sigma_circle[1, :], 'y', label='Σ·Vᵀ·x')
plt.plot(U_circle[0, :], U_circle[1, :], 'b', label='U·Σ·Vᵀ·x')
plt.axis('equal')
plt.legend()
plt.title("SVD Geometric Interpretation (2×2 matrix)")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
